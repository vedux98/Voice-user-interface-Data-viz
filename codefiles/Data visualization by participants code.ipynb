```python
import pandas as pd 
import seaborn as sns
import  matplotlib.pyplot as plt 
import numpy as np 
```

```python
df = pd.read_excel("Aggregated data.xlsx")
df
```

```python
color_mapping = {
    'Acoustic cues': 'yellow',
    'Linguistic cues': 'green',
    'Other cues': 'blue'
}
color_matrix = df.replace(color_mapping)
```

```python
datamapping = {
    
    "Acoustic cues":1,
    "Linguistic cues":2,
    "Other cues": 3
}
```

```python
df[['Situation 1', 'Situation 2', 'Situation 3', 'Situation 4']] = df[['Situation 1', 'Situation 2', 'Situation 3', 'Situation 4']].replace(datamapping)
```

```python
df
```

```python
df.fillna("")
```

```python
fig, ax = plt.subplots(figsize=(10, 20))  
ax.set_xticks(np.arange(len(df.columns)))
ax.set_yticks(np.arange(len(df)))

ax.set_xticklabels(df.columns)
ax.set_yticklabels(df.index)

for i in range(len(df)):
    for j in range(len(df.columns)):
        ax.fill(j, i, color_matrix.iloc[i, j])
ax.set_title("User responses to situation")
plt.xlabel("Situations")
plt.ylabel("Participants")
plt.show()
```

```python
colors = ['yellow', 'green', 'blue']
cmap = plt.cm.colors.ListedColormap(colors)

fig, ax = plt.subplots(figsize=(10, 20))
cax = ax.imshow(df, cmap=cmap, aspect='auto')

ax.set_xticks(np.arange(df.shape[1]+1)-.5, minor=True)
ax.set_yticks(np.arange(df.shape[0]+1)-.5, minor=True)
ax.grid(which="minor", color="w", linestyle='-', linewidth=2)
ax.tick_params(which="minor", size=0)


# Adjust the appearance of the plot
ax.set_xticks(range(df.shape[1]))
ax.set_yticks(range(df.shape[0]))
ax.set_xticklabels(df.columns, rotation=45, ha='right')
ax.set_yticklabels(range(1, df.shape[0] + 1)) 
ax.set_xlabel('Situations')
ax.set_ylabel('Participants')
ax.set_title('User responses to situation')
cbar = plt.colorbar(cax, ticks=[1, 2, 3], orientation='vertical')
cbar.ax.set_yticklabels(['Acoustic cues', 'Linguistic cues', 'Other cues']) 

plt.tight_layout()
plt.show()
```
